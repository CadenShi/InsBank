```python
from scipy.stats import spearmanr

for dataset in datasets:
    print(f"Processing {dataset['name']}")

    data = torch.load(dataset['path'])
    complexity = []
    quality = []
    instruction_len = []
    response_len = []
    for sample in data:
        # if sample['complexity'] > 3.5:
        #     continue
        flag1 = False
        flag2 = False
        tmp_instruction_len = []
        tmp_response_len = []
        conversations = sample['conversations']
        for turn in conversations:
            if turn['from'] == 'system':
                continue
            if turn['from'] in ['human', 'user']:
                tmp_instruction_len.append(len(turn['value'].strip().split(' ')))
                flag2 = True
            else:
                tmp_response_len.append(len(turn['value'].strip().split(' ')))
                flag1 = True
        if flag1 and flag2:
            complexity.append(sample['complexity'])
            quality.append(sample['quality'])
            instruction_len.append(sum(tmp_instruction_len)/len(tmp_instruction_len))
            response_len.append(sum(tmp_response_len)/len(tmp_response_len))
    if len(complexity) < 1:
        continue

    print(f'Examples: {len(complexity)}')
    print(f'Complexity: {sum(complexity)/len(complexity)}')
    print(f'Quality: {sum(quality)/len(quality)}')
    print(f'Instruction Len: {sum(instruction_len)/len(instruction_len)}')
    print(f'Response Len: {sum(response_len)/len(response_len)}')
    print(f'Complexity - Ins Len: {spearmanr(complexity, instruction_len)}')
    print(f'Complexity - Response Len: {spearmanr(complexity, response_len)}')
    print(f'Quality - Response Len: {spearmanr(quality, response_len)}')
    print(f'Quality - Complexity: {spearmanr(quality, complexity)}')
    print(f'')

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

for dataset in datasets:
    print(f"Processing {dataset['name']}")
    data = torch.load(dataset['path'])
    complexity = [item['complexity'] for item in data]
    quality = [item['quality'] for item in data]

    data1 = np.array(quality)
    data2 = np.array(complexity)

    sns.set(style="whitegrid")

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    name = dataset['name']

    sns.histplot(data1, bins=30, kde=True, color='blue', stat='density', ax=axs[0])
    axs[0].set_title(f'Frequency Distribution of {name} Quality', fontsize=16)
    axs[0].set_xlabel('Value', fontsize=14)
    axs[0].set_ylabel('Density', fontsize=14)
    axs[0].set_xlim(1, 6)

    sns.histplot(data2, bins=30, kde=True, color='orange', stat='density', ax=axs[1])
    axs[1].set_title(f'Frequency Distribution of {name} Complexity', fontsize=16)
    axs[1].set_xlabel('Value', fontsize=14)
    axs[1].set_ylabel('Density', fontsize=14)
    axs[1].set_xlim(1, 6)

    plt.tight_layout()
    plt.show()

```

```python
import os

import json
import random
from tqdm import tqdm


def j_load(path):
    if path.endswith('.jsonl'):
        data_list = []
        with open(path, 'r', encoding='utf-8') as file:
            for line in file:
                try:
                    data_list.append(json.loads(line))
                except:
                    continue
        return data_list
    else:
        with open(path, 'r') as f:
            return json.load(f)

mt_bench = j_load('/mnt/public02/usr/yuanpeiwen/instruction_datasets/mt_bench/question.jsonl')
```

```python
first_turn = []
second_turn = []
for item in mt_bench:
    first_turn.append(item['turns'][0])
    second_turn.append(item['turns'][1])

first_turn_format = []
for question in first_turn:
    sample = {
        'instruction': question,
    }
    first_turn_format.append(sample)

second_turn_format = []
for question in second_turn:
    sample = {
        'instruction': question,
    }
    second_turn_format.append(sample)
```

```python
with open('.._cleaned/analysis/mtbench_second_turn.json', 'w') as f:
    json.dump(second_turn_format, f, indent=2)

with open('.._cleaned/analysis/mtbench_first_turn.json', 'w') as f:
    json.dump(first_turn_format, f, indent=2)
```
